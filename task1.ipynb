# Аттрактатор Лоренца

Выполнила Петрова Наталия

\begin{align}
\dot{x} & = \sigma(y - x) \\
\dot{y} & = x(r - z) - y \\
\dot{z} & = xy - bz \\
\end{align}

```python
import numpy as np
from numpy import linspace, array
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D
from scipy.integrate import solve_ivp
```

```python
"""
Params
"""
sigma = 10.0
r = 28.0
b = 2.667

"""
Initial conditions
"""
init = [0, 1, 0]

"""
Lorenz Attractor
"""
def lorenz(t, xyz):
    x, y, z = xyz
    x_dot = sigma*(y - x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return x_dot, y_dot, z_dot

t_span = [0, 50]
solution = solve_ivp(lorenz, t_span, init, t_eval = np.linspace(0, 50, 100000))

fig = plt.figure()
ax = Axes3D(fig)

ax.plot(solution.y[0], solution.y[1], solution.y[2], 'green')
ax.set_title("Lorenz Attractor")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```
